```python
def Trayectoria(v0, tetha, x, y0, g = 9.807):
  y = x * m.tan(m.radians(tetha)) - ((g * x**2)/(2*v0**2 * m.cos(tetha)**2))+y0
  return y
```

```python
promedio = [0]*12
for i in range(12):
  valor_promedio = 0
  for ao in Oxford_sun_hours:
    valor_promedio = int(valor_promedio + ao[i])
    promedio[i] = valor_promedio
promedio
```

```python
mayor = 0
for i,mes in enumerate(promedio, 1):
    if mes > mayor:
      mayor = mes
      month = i

mayor,month
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize =(20,15))

for i in range(12):
  MesData = [ao[i] for ao in Oxford_sun_hours]
  plt.subplot (3, 4, i+1)
  plt.hist(MesData, bins =10)
  plt.title(f"Histograma del Mes {i+1}")
  plt.xlabel ('Horas de sol')
  plt.ylabel ('Frecuencia')

plt.tight_layout()
plt.savefig("Grfico_histogramas.png")
plt.show()
```

```python
Americas = pd.read_csv("https://raw.githubusercontent.com/robintux/Datasets4StackOverFlowQuestions/master/AmericaSur.csv",
                       index_col = 0)
```

```python
Americas.replace("NA",np.nan , inplace=True)
```

```python
plt.figure(figsize=(20,15))

for i,pais in enumerate(Americas,1):
  plt.subplot(3, 4 , i)
  plt.scatter(Americas.index, Americas[pais])
  plt.title(f"Grficos de dispersin de {pais}")
  plt.xlabel("Aos")
  plt.ylabel("valor")
  plt.grid(True)

plt.tight_layout()
plt.show()

```
